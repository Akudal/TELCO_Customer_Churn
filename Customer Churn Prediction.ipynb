# Building a Prediction Model

```python
#Importing necessary libraries

!pip install imbalanced-learn
import pandas as pd
import sklearn
from sklearn import metrics
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeClassifier
import imblearn
from imblearn.combine import SMOTEENN
```

```python
df = pd.read_csv(r'tel_churn.csv')
pd.set_option('display.max_columns', None)

df.head()
```

```python
df.columns
```

```python
df = df.drop('Unnamed: 0', axis = 1)
df.head()
```

```python
#Creating Independent (X) and Dependent (y) variables
X = df.drop('Churn', axis = 1)
X.head()
```

```python
y = df['Churn']
y.head()
```

```python
print(X.shape)
print()
print(y.shape)
```

```python
#Bifurcating the Variables into train and test variables
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
# test_size = 0.2 means 20% of rows will be considered as test data and 80% as training data
```

### Decision Tree Classifier

```python
model_dt = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8)
```

```python
model_dt.fit(X_train, y_train)
```

```python
#This algorithm basically predicts Churn (y) using a prediction model that is trained from all Independent variables (X)
y_pred = model_dt.predict(X_test)
```

```python
y_pred
```

```python
model_dt.score(X_test, y_test)
```

```python
print(classification_report(y_test, y_pred, labels=[0,1]))
```

**After looking at the minority class i.e., 1 (Churners); The model has low percision and f1-scores due to imbalance in dataset.**

```python
print(confusion_matrix(y_test, y_pred))
```

### Leveraging SMOTEENN Analysis for Upsampling (SMOTE) and Cleaning (ENN)

```python
!pip install -U imbalanced-learn
```

```python
from imblearn.combine import SMOTEENN
```

```python
sm = SMOTEENN()
X_resampled, y_resampled = sm.fit_resample(X, y)
```

```python
X_r_train, X_r_test, y_r_train, y_r_test = train_test_split(X_resampled, y_resampled, test_size = 0.2)
```

```python
model_dt_smote = DecisionTreeClassifier(criterion = 'gini', random_state = 100, max_depth = 6, min_samples_leaf = 8)
```

```python
model_dt_smote.fit(X_r_train, y_r_train)
```

```python
y_pred_smote = model_dt_smote.predict(X_r_test)
```

```python
print(classification_report(y_r_test, y_pred_smote, labels=[0,1]))
```

```python
model_dt_smote.score(X_r_test, y_r_test)
```

```python
print(confusion_matrix(y_r_test, y_pred_smote))
```

**There is a significant improvement in the precision, recall and f1 scores. This indicates that the prediction model has high accuracy.**

**After looking at the minority class i.e., 1 (Churners); The model has low percision and f1-scores due to imbalance in dataset.**
